```python
import cv2
import numpy as np
from math import log
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('th.jpg', 0)
cv2_imshow(image)
cv2.waitKey(0)#<---0 tells us that image will close upon pressing any key
cv2.destroyAllWindows()
```

```python
c=255/np.log(1+np.max(image))
newi=c*np.log(image*1)
newi=np.array(newi,dtype=np.uint8)
cv2_imshow(newi)
```

```python
g = [0.2,0.5,1.2,1.8,3.2]
for i in g:
        newImg = 255 * np.power((image/255),i)
        print(f'Gamma Value: {i}')
        cv2_imshow(newImg)
        print('\n')
```

```python
import pandas as pd
[w,h] = image.shape
for i in range(0,w):
  for j in range(0,h):
    if newImg[i][j] > 100 and newImg[i][j]<200:
      newImg[i][j] = 210
cv2_imshow(newImg)
```
