```python
import cv2
import numpy as np
from math import log
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('th.jpg', 0)
cv2_imshow(image)
cv2.waitKey(0)#<---0 tells us that image will close upon pressing any key
cv2.destroyAllWindows()
```

gamma **transformation**

```python
c=255/np.log(1+np.max(image))
newi=c*np.log(image*1)
newi=np.array(newi,dtype=np.uint8)
cv2_imshow(newi)
```

```python
g = [0.2,0.5,1.2,1.8,3.2]
for i in g:
        newImg = 255 * np.power((image/255),i)
        print(f'Gamma Value: {i}')
        cv2_imshow(newImg)
        print('\n')
```

gray level **scaling** 

```python
import pandas as pd
[w,h] = image.shape
for i in range(0,w):
  for j in range(0,h):
    if newImg[i][j] > 100 and newImg[i][j]<200:
      newImg[i][j] = 210
cv2_imshow(newImg)
```

```python
bit plane slicing
```

```python
[w,h]=image.shape
img1 = np.zeros([w,h,3], dtype=np.uint8)
img2 = np.zeros([w,h,3], dtype=np.uint8)
img3 = np.zeros([w,h,3], dtype=np.uint8)
img4 = np.zeros([w,h,3], dtype=np.uint8)
img5 = np.zeros([w,h,3], dtype=np.uint8)
img6 = np.zeros([w,h,3], dtype=np.uint8)
img7 = np.zeros([w,h,3], dtype=np.uint8)
img8 = np.zeros([w,h,3], dtype=np.uint8)

def bitget(nbr,pos):
  return (nbr >> pos) & 1

planes = 8
for i in range(0, w):
  for j in range(0, h):
    for p in range(planes - 1, -1, -1):
      if p == 7:
        img1[i][j] = 255 * bitget(image[i][j], p)
      elif p == 6:
        img2[i][j] = 255 * bitget(image[i][j], p)
      elif p == 5:
        img3[i][j] = 255 * bitget(image[i][j], p)
      elif p == 4:
        img4[i][j] = 255 * bitget(image[i][j], p)
      elif p == 3:
        img5[i][j] = 255 * bitget(image[i][j], p)
      elif p == 2:
        img6[i][j] = 255 * bitget(image[i][j], p)
      elif p == 1:
        img7[i][j] = 255 * bitget(image[i][j], p)
      else:
        img8[i][j] = 255 * bitget(image[i][j], p)
  
  cv2_imshow(img1)
  cv2_imshow(img2)
  cv2_imshow(img3)
  cv2_imshow(img4)
  cv2_imshow(img5)
  cv2_imshow(img6)
  cv2_imshow(img7)
  cv2_imshow(img8)

```
