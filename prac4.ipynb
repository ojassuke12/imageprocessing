```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.jpg',0)
histr=cv2.calcHist([img],[0],None,[256],[0,256])
plt.title("Image")
plt.xlabel('bins')
plt.ylabel("No. of pixels")
plt.hist(histr,density=True,bins=40)
plt.show()
```
